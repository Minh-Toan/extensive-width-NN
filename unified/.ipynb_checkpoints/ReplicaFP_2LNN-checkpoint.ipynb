```python
def Ez(f):
    integrand = lambda z: f(z)*np.exp(-z**2/2)/np.sqrt(2*np.pi)
    return quad(integrand, -10, 10)[0]

def coeffs(f):
    c0 = Ez(lambda z: f(z))
    c1 = Ez(lambda z: z*f(z))
    c2 = Ez(lambda z: (z**2-1)*f(z))
    nu = Ez(lambda z: f(z)**2)
    return c0, c1, c2, nu

# Choose the readout type
# Options: 'Read_gauss', 'Read_ones', 'Read_unif', 'Read_rad', Read_4point
readouts='Read_gauss'

# Load the CSV file (assuming comma-separated values)
# The file is assumed to have three rows:
#   1. lambdas
#   2. m(lambda) (where m = 1 - mmse)
#   3. integral of m from 0 to lambda: M(lambda) = ∫0^λ m(t) dt
if readouts=='Read_gauss':
  data = np.loadtxt('gauss.csv', delimiter=',')
elif readouts=='Read_rad':
  data = np.loadtxt('rad.csv', delimiter=',')
elif readouts=='Read_ones':
  data = np.loadtxt('ones.csv', delimiter=',')
elif readouts=='Read_unif':
  data = np.loadtxt('unif.csv', delimiter=',')
elif readouts=='Read_4point':
  data = np.loadtxt('4point.csv', delimiter=',')

# Extract data from the rows
lambdas = data[0, :]
m_values = data[1, :]
M_values = data[2, :]

# Create spline interpolations for m(λ) and its integral M(λ)
# using cubic splines. The 'extrapolate' option ensures values
# are given outside the interpolation range if needed.
m_spline = CubicSpline(lambdas, m_values, extrapolate=True)
M_spline = CubicSpline(lambdas, M_values, extrapolate=True)

# Define the mmse function using the relation: mmse = 1 - m
def m_05(lam):
    return m_spline(lam).item()

# Define the integrated mmse function:
# since mmse(λ)=1-m(λ), its integral is:
#  ∫_0^λ mmse(t) dt = λ - ∫_0^λ m(t) dt = λ - M(λ)
def iota_05(lam):
    return (lam - M_spline(lam))/4


#-----------------------------------------------------------
# The following functions are used to compute the free entropy and mmse for the universal phase, i.e. ONLY Q_W=0
#-----------------------------------------------------------    

def Ez(f):
    integrand = lambda z: f(z)*np.exp(-z**2/2)/np.sqrt(2*np.pi)
    return quad(integrand, -10, 10)[0]

def coeffs(f):
    c0 = Ez(lambda z: f(z))
    c1 = Ez(lambda z: z*f(z))
    c2 = Ez(lambda z: (z**2-1)*f(z))
    nu = Ez(lambda z: f(z)**2)
    return c0, c1, c2, nu

def q_2_05(alpha, Delta, phi): # overlap of S
    c0, c1, c2, nu = coeffs(phi)
    Delta_2 = (Delta + nu - c0**2 - c1**2 - c2**2/2)/c2**2*2
    def eq(r):
        return m_05(r) - 1 - Delta_2 + 2*alpha/r
    init = 2*alpha/(0.5+Delta_2)
    r = fsolve(eq, init)[0]
    return 1 + Delta_2 - 2*alpha/r

def q_2_and_hat_q_2_05(alpha, Delta, phi): # overlap of S and hat variable
    c0, c1, c2, nu = coeffs(phi)
    Delta_2 = (Delta + nu - c0**2 - c1**2 - c2**2/2)/c2**2*2
    def eq(r):
        return m_05(r) - 1 - Delta_2 + 2*alpha/r
    init = 2*alpha/(0.5+Delta_2)
    r = fsolve(eq, init)[0]
    return 1 + Delta_2 - 2*alpha/r, r

def free_entropy_univ_05(alpha, delta,phi): # free entropy of the universal phase
  c0, c1, c2, nu = coeffs(phi)
  if (c2 == 0):
    q2 = 0
    hat_q2 = 0
    return - np.log(2*np.pi)/2 - (np.log(delta + c2**2*(1-q2)/2 +g(1) - g(0)))/2 - 1/2
  else:
    q2, hat_q2 = q_2_and_hat_q_2_05(alpha, delta, phi)
    return  - np.log(2*np.pi)/2 - (np.log(delta + c2**2*(1-q2)/2 +g(1) - g(0)))/2 - 1/2 + hat_q2*(1-q2)/(4*alpha) - iota_05(hat_q2)/alpha          
```

```python
#-----------------------------------------------------------
# The following functions are used to compute the free entropy and mmse for the full phase space (universal and specialised) phase, i.e. Q_W=0 and Q_W>0
#-----------------------------------------------------------

# Chose between rademacher or gaussian theory

#opt = 'rad'
opt = 'gauss'

def psi_w(l):
  if opt == 'rad':
    # overlap function, rademacher prior
    integrand = lambda z: tanh(sqrt(l)*z + l)*exp(-z**2/2)/sqrt(2*pi)
    return quad(integrand, -10, 10)[0]
  elif opt == 'gauss':
    # overlap function, gaussian prior
    return l/(1+l)

def m_inv(q):
  if opt == 'rad':
    # inverse overlap function, rademacher prior  
    u = 1
    if psi_w(u)>=q:
         r = bisect(lambda x: psi_w(x)-q, 0, 1)
    else:
        while psi_w(u)<q:
            u *= 2
        r = bisect(lambda x: psi_w(x)-q, u/2, u)
    return r
  elif opt == 'gauss':
    # inverse overlap function, gaussian prior
    return q/(1-q)  

def psi(g):
  if opt == 'rad':
    # free energy function, rademacher prior
    integrand = lambda z: log( cosh( sqrt(g)*z + g ) )*exp( -z**2/2 )/sqrt(2*pi)
    return -g/2 + quad(integrand, -10, 10)[0]
  elif opt == 'gauss':
    # free energy function, gaussian prior
    return g/2 - log(g+1)/2

def EE_v2_g(q_W,v,P_v,g): # returns the expectation wrt P_v of the function v^2 g(q_W(v))
  exp_v2_g = 0 
  for t in range(v.shape[0]):
    exp_v2_g += P_v[t]*v[t]**2*g(q_W[t])
  return exp_v2_g

def EE_v2_qW2(q_W,v,P_v): # returns the expectation wrt P_v of the function v^2 q_W(v)^2
  exp_v2_qW2 = 0
  for t in range(v.shape[0]):
    exp_v2_qW2 += P_v[t]*v[t]**2*q_W[t]**2
  return exp_v2_qW2 

def EE_v(arr,P_v): # returns the expectation wrt P_v of the function arr
  exp_arr = 0
  for t in range(P_v.shape[0]):
    exp_arr += P_v[t]*arr[t]
  return exp_arr

def m_s_inv_05(q_W, gamma, v, P_v):
    # Inverse overlap function: find l such that m(gamma, l) equals (1 - q_W**2) / 4.
    target = EE_v2_qW2(q_W, v, P_v)
    def eq(l_arr):
        l_val = l_arr[0]
        m_val = m_05([l_val])
        if not np.isfinite(m_val):
            return 1e8
        return m_val - target

    # fsolve requires the initial guess to be an array.
    l_initial = np.array([0.00001])
    l_solution = fsolve(eq, l_initial)
    return l_solution[0]

def iota_diff_05(r,t):
# mutual information of the channel sqrt(l)*W.T Dv W/sqrt(n) + Z
# W Gaussian size nxd, n/d = ratio
# Z Gaussian symmetric
  return iota_05(r+t) - iota_05(t)

def Kappa_full_NSB_05(q_W,v,P_v,g,coeff,alpha,gamma,delta,rho_2): # overlap of S
    r_K = coeff[1]**2 + rho_2*coeff[2]**2/2 + g(1)
    q_K_new = r_K/2
    #print('r_K =',r_K)
    #print('-----------------')
    for i in range(100):
        r_2 = alpha*(1/(delta+r_K-q_K_new))*(coeff[2]**2)
        tilde_q = m_s_inv_05(q_W,gamma,v,P_v)
        q_2 = m_05(np.array([tilde_q + r_2]))+(rho_2-1)
        q_K_old = q_K_new
        q_K_new = coeff[1]**2 + coeff[2]**2/2*q_2 + EE_v2_g(q_W,v,P_v,g)
        #print('q_K =',q_K_new)
        if abs(q_K_new - q_K_old)<1e-5:
            #print('-----------------')
            break
    return r_K,q_K_new

def free_entropy_full_NSB_05(q_W,v,P_v,g,g_prime,coeff,alpha,gamma,delta,rho_2): # another way to compute the potiential in W-overlap
     r_W = np.zeros(v.shape[0])
     psi_rW = np.zeros(v.shape[0])
     r_K,q_K = Kappa_full_NSB_05(q_W,v,P_v,g,coeff,alpha,gamma,delta,rho_2)
     r_2 = alpha*(1/(delta+r_K-q_K))*(coeff[2]**2)
     tilde_q = m_s_inv_05(q_W,gamma,v,P_v)
     q_2 = m_05(np.array([tilde_q + r_2]))+(rho_2-1)
     for t in range(v.shape[0]):
       r_W[t] = 1/(2*gamma)*(q_2 - (rho_2-1) - EE_v2_qW2(q_W,v,P_v))*(-2*v[t]**2*q_W[t]/derivative_mmse_05(tilde_q, gamma)) + alpha/gamma*v[t]**2*g_prime(q_W[t])/(delta + r_K - q_K)
       psi_rW[t] = psi(r_W[t])
     # compute the potential
     return (-q_2*r_2/2 - alpha*np.log((delta + r_K - q_K)) + rho_2*r_2/2 - EE_v(gamma*q_W*r_W - 2*gamma*psi_rW,P_v) - 2*iota_diff_05(r_2,tilde_q) - alpha)/(2*alpha) - np.log(2*np.pi)/2

def derivative_mmse_05(l, ratio, h=1e-6):
    #Numerically compute the derivative of mmse_MF with respect to l
    #l (float): Point at which to compute the derivative.
    #ratio (float): The ratio parameter passed to mmse_MF.
    #h (float): A small perturbation for the finite difference.
    f_plus = 1 - m_05(l + h)
    f_minus = 1 - m_05(l - h)
    return (f_plus - f_minus) / (2 * h)         

def mmse_qW_q2_full_NSB_iter_05(q_W_init,v,P_v,g,g_prime,coeff,alpha,gamma,delta,rho_2): # returns mmse and overlaps computed iteratively
  dump=1
  q_W_iter = np.copy(q_W_init)
  r_W = np.zeros(v.shape[0])
  q_W_old = np.zeros(v.shape[0])  
  for i in range(500):
     r_K,q_K = Kappa_full_NSB_05(q_W_iter,v,P_v,g,coeff,alpha,gamma,delta,rho_2)
     r_2 = alpha*(1/(delta+r_K-q_K))*(coeff[2]**2)
     tilde_q = m_s_inv_05(q_W_iter,gamma,v,P_v)
     q_2 = m_05(np.array([tilde_q + r_2]))+(rho_2-1)

     for t in range(v.shape[0]):
       r_W[t] = 1/(2*gamma)*(q_2 - (rho_2-1) - EE_v2_qW2(q_W_iter,v,P_v))*(-2*v[t]**2*q_W_iter[t]/derivative_mmse_05(tilde_q, gamma)) + alpha/gamma*v[t]**2*g_prime(q_W_iter[t])/(delta + r_K - q_K)
       q_W_old[t] = q_W_iter[t]
       q_W_iter[t] = psi_w(r_W[t])
       q_W_iter[t] = dump*q_W_iter[t] + (1-dump)*q_W_old[t]
     if np.abs(q_W_iter - q_W_old).all()<1e-7:
          #print('-----------------')
          break
  return r_K-q_K, q_W_iter, q_2      

#-----------------------------------------------------------
# The following functions are used to compute the free entropy and mmse for the old sp phase (Kabashima-Sakata ansatz)
#-----------------------------------------------------------

def Kappa_NSB(q_W,v,P_v,g,coeff,alpha,gamma,delta,rho_2): # compute Kappa
  r_K = coeff[1]**2 + rho_2*coeff[2]**2/2 + g(1)
  q_K_new = r_K/2
  for i in range(100):
      r_2 = alpha*(1/(delta+r_K-q_K_new))*(coeff[2]**2)
      q_2 = rho_2 - (1 - EE_v2_qW2(q_W,v,P_v))/(1 + r_2*(1 - EE_v2_qW2(q_W,v,P_v)))
      q_K_old = q_K_new
      q_K_new = coeff[1]**2 + coeff[2]**2/2*q_2 + EE_v2_g(q_W,v,P_v,g) 
      #print(q_K_new)
      if abs(q_K_new - q_K_old)<1e-5:
          #print('-----------------')
          break    
  return q_K_new,r_K

def mmse_qW_q2_NSB(q_W_init,v,P_v,g,g_prime,coeff,alpha,gamma,delta,rho_2): # returns mmse and overlaps computed iteratively
  damp = 1
  q_W_iter = np.copy(q_W_init)
  r_W = np.zeros(v.shape[0])
  q_W_old = np.zeros(v.shape[0])
  for i in range(500):
     q_K,r_K = Kappa_NSB(q_W_iter,v,P_v,g,coeff,alpha,gamma,delta,rho_2)
     r_2 = alpha*(1/(delta+r_K-q_K))*(coeff[2]**2)
     q_2 = rho_2 - (1 - EE_v2_qW2(q_W_iter,v,P_v))/(1 + r_2*(1 - EE_v2_qW2(q_W_iter,v,P_v)))
     for t in range(v.shape[0]):
       r_W[t] = r_2*v[t]**2*q_W_iter[t]/gamma/(1 + r_2*(1 - EE_v2_qW2(q_W_iter,v,P_v))) + alpha/gamma*v[t]**2*g_prime(q_W_iter[t])/(delta + r_K - q_K)
       q_W_old[t] = q_W_iter[t]
       q_W_iter[t] = psi_w(r_W[t])
       q_W_iter[t] = (1-damp)*q_W_old[t] + damp*q_W_iter[t]
     if np.abs(q_W_iter - q_W_old).all()<1e-7:
          #print('-----------------')
          break
  return r_K-q_K, q_W_iter, q_2

def free_entropy_NSB(q_W,v,P_v,g,g_prime,coeff,alpha,gamma,delta,rho_2): 
    psi_rW = np.zeros(v.shape[0])
    r_W = np.zeros(v.shape[0])
    q_K,r_K = Kappa_NSB(q_W,v,P_v,g,coeff,alpha,gamma,delta,rho_2)
    r_2 = alpha*(1/(delta+r_K-q_K))*(coeff[2]**2)
    q_2 = rho_2 - (1 - EE_v2_qW2(q_W,v,P_v))/(1 + r_2*(1 - EE_v2_qW2(q_W,v,P_v)))
    for t in range(v.shape[0]):
       r_W[t] = r_2*v[t]**2*q_W[t]/gamma/(1 + r_2*(1 - EE_v2_qW2(q_W,v,P_v))) + alpha/gamma*v[t]**2*g_prime(q_W[t])/(delta + r_K - q_K)
       psi_rW[t] = psi(r_W[t])
    return (-q_2*r_2/2 - alpha*np.log((delta + r_K - q_K)) + rho_2*r_2/2 - EE_v(gamma*q_W*r_W - 2*gamma*psi_rW,P_v) - (1/2)*log(1 + r_2*((1 - EE_v2_qW2(q_W,v,P_v)))) - alpha)/(2*alpha) - np.log(2*np.pi)/2
```

```python
def hermite_coefficients(func, n, a, b):
    """
    Computes the first n Hermite coefficients of a function.

    Args:
        func: The function whose coefficients are to be computed.
        n: The number of coefficients to compute.
        a: The lower limit of the integration interval.
        b: The upper limit of the integration interval.

    Returns:
        A NumPy array containing the first n Hermite coefficients.
    """

    # Define the Hermite polynomials
    def hermite(x, n):
        if n == 0:
            return 1
        elif n == 1:
            return x
        else:
            return x * hermite(x, n - 1) - (n - 1) * hermite(x, n - 2)

    # Define the weight function
    def weight(x):
        return np.exp(-x**2/2)/np.sqrt(2*np.pi)

    # Compute the coefficients
    coefficients = []
    for i in range(n+1):
        # Use scipy.integrate.quad to compute the integral numerically.
        # Note: The integrand needs to be defined as a lambda function.
        integral, error = scipy.integrate.quad(lambda x: func(x) * hermite(x, i) * weight(x), a, b)
        coefficients.append(integral)

    return np.array(coefficients)
```

```python
def get_states_and_probs(num_v):
    # For an equiprobable, zero mean, unit variance distribution,
    # we choose symmetric values and then scale them appropriately.
    if num_v % 2 == 1:
        # For odd num_v, use integers symmetric about zero.
        values = np.arange(-num_v // 2, num_v // 2 + 1)
    else:
        # For even num_v, use half-integer offsets.
        values = np.arange(-num_v / 2 + 0.5, num_v / 2 + 0.5, 1)
    
    # Compute the current variance (mean of squares) of these values.
    current_variance = np.mean(values**2)
    
    # Scale the values so that the new variance becomes 1.
    states = values / np.sqrt(current_variance)
    
    # Since states are equiprobable, each probability is 1/num_v.
    probs = np.ones(num_v) / num_v
    
    return states, probs

def get_discrete_gaussian_on_grid(num_v, mu, sigma, mult_sigma):
    # Create a grid of states between -mult*sigma and mult*sigma around the mean mu.
    v = np.linspace(mu-mult_sigma*sigma, mu+mult_sigma*sigma, num_v)
    # Compute unnormalized probabilities using the Gaussian pdf with mean mu and variance sigma
    weights = np.exp(-((v-mu)**2) / (2*sigma**2)) / np.sqrt(2 * np.pi * sigma**2)
    # Normalize the weights to get probabilities
    P_v = weights / np.sum(weights)
    return v, P_v    
```

```python
# choose the activation function among the activations below
# the cell computes the first num_coeff Hermite coefficients and defines the function g()

activation = 'Quadratic + Cubic' # 'ReLU', 'Tanh', 'Softplus', 'Erf', 'Arctan', 'LeakyReLU', 'ELU', 'Sigmoid', 'Quadratic', 'Cubic', 'Quadratic + Cubic', 'Linear + Quadratic + Cubic', 'Quadratic + Quartic'
num_coeff = 15

if (activation == 'ReLU'):
  analytic = True
else:
  analytic = False


if (analytic == True):
 if (activation == 'ReLU'):

      def sigma(x):
        return np.heaviside(x,1)*x - 1/np.sqrt(2*np.pi)

      def g(x):
        return -1/(2*np.pi) - x**2/(4*np.pi) + np.sqrt(1-x**2)/(2*np.pi) + x*np.arctan(x/np.sqrt(1-x**2))/(2*np.pi)

      def g_prime(x):
        return (np.arctan(x/np.sqrt(1-x**2))-x)/(2*np.pi)

      coefficients = hermite_coefficients(lambda x: sigma(x), num_coeff, -10, 10)

else:

  if (activation == 'ReLU_approx'):
    def sigma(x):
        return np.heaviside(x,1)*x - 1/np.sqrt(2*np.pi)

  if (activation == 'Tanh'):
    def sigma(x):
        return np.tanh(2*x)
     
  if (activation == 'Softplus'):
    def sigma(x):
        return np.log(1+np.exp(x))  

  if (activation == 'Erf'):
    def sigma(x):
        return 0.5*(1 + special.erf(x/np.sqrt(2)))

  if (activation == 'Arctan'):
    def sigma(x):
        return np.arctan(x)

  elif (activation == 'LeakyReLU'):
    def sigma(x):
        return np.heaviside(x,1)*x + 0.3*np.heaviside(-x,-1)*x

  elif (activation == 'ELU'):
    def sigma(x):
        return np.heaviside(x,1)*x + np.heaviside(-x,1)*(np.exp(x) - 1)

  elif (activation == 'Sigmoid'):
    def sigma(x):
        return 1/(1+np.exp(-x))

  elif (activation == 'Quadratic'):
    def sigma(x):
        return np.power(x,2)/np.sqrt(2)

  elif (activation == 'Cubic'):
    def sigma(x):
        return (x**3 - 3*x)/np.sqrt(6)

  elif (activation == 'Quadratic + Cubic'):
    def sigma(x):
        return (x**2 - 1)/np.sqrt(2) + (x**3 - 3*x)/6

  elif (activation == 'Linear + Quadratic + Cubic'):
    def sigma(x):
        return x + (x**2 - 1)/np.sqrt(2) + (x**3 - 3*x)/6

  elif (activation == 'Quadratic + Quartic'):
    def sigma(x):
        return (x**2 - 1)/np.sqrt(2) + (x**4 - 6*x**2 + 3)/8

  coefficients = hermite_coefficients(lambda x: sigma(x), num_coeff, -10, 10)

  # Create the approximation function using the computed coefficients
  def g(x):
    g = 0
    for i in range(3,len(coefficients)):
      g += coefficients[i]**2 * x**i / math.factorial(i)
    return g

  def g_prime(x):
    g = 0
    for i in range(3,len(coefficients)):
      g += coefficients[i]**2 * x**(i-1) / math.factorial(i-1)
    return g

# Plot the original function
x = np.linspace(-2, 2, 100)
plt.plot(x, sigma(x), label='activation function')

plt.xlabel('x')
plt.ylabel('sigma(x)')
plt.legend()
plt.grid(True)
plt.show()

print(coefficients)
```
