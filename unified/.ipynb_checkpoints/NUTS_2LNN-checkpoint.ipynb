```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist
from numpyro.infer import MCMC, NUTS
import numpy as np
import math
from jax import random

jax.config.update("jax_enable_x64", True)

# Set random keys
rng_key = jax.random.PRNGKey(0)
rng_key, key_teacher, key_data = jax.random.split(rng_key, 3)

# Hyperparameters
d = 180                 # Input dimension (k0)
gamma1 = 0.5            # k1/d
#gamma2 = 0.5            # k2/d
alpha = 2.5             # n/d^2
k1 = int(gamma1*d)      # First hidden layer
#k2 = int(gamma2*d)      # Second hidden layer
n = int(alpha*d**2)     # Number of samples
delta = 1.25             # Noise variance
sigma = jnp.sqrt(delta) # Noise standard deviation
info_init = 1           # Info/Uninfo initialization

num_warmup_samples = 50
num_posterior_samples = 20

def activation(x):
    return (x**2 - 1)/np.sqrt(2) + (x**3 - 3*x)/6

# Teacher network parameters
W1_teacher = jax.random.normal(key_teacher, (k1, d), dtype=jnp.float64)
#W2_teacher = jax.random.normal(jax.random.split(key_teacher)[0], (k2, k1), dtype=jnp.float64)

# Generate the readouts v_teacher
readouts = 'ones'
if readouts == 'gaussian':
    v_teacher = jax.random.normal(key_teacher, (k1,), dtype=jnp.float64)
elif readouts == 'uniform':
    v_teacher = jax.random.uniform(key_teacher, (k1,), minval=-np.sqrt(3), maxval=np.sqrt(3), dtype=jnp.float64)
elif readouts == 'rademacher':
    v_teacher = np.sign(jax.random.normal(jax.random.split(key_teacher)[1], (k1,), dtype=jnp.float64))    
elif raedouts == 'ones':
    v_teacher = np.ones(k1)
elif readouts == 'ternary':
    v_teacher = random.choice(key_teacher, jnp.array([-jnp.sqrt(3/2), 0.0, jnp.sqrt(3/2)]), shape=(k1,), p=jnp.array([1/3, 1/3, 1/3]))
elif readouts == 'quaternary':
    v_teacher = random.choice(key_teacher, jnp.array([-3/jnp.sqrt(5), -1/jnp.sqrt(5), 1/jnp.sqrt(5), 3/jnp.sqrt(5)]), shape=(k1,), p=jnp.array([1/4, 1/4, 1/4, 1/4]))
elif readouts == 'quinary':
    v_teacher = random.choice(key_teacher, jnp.array([-jnp.sqrt(2), -1/jnp.sqrt(2), 0.0, 1/jnp.sqrt(2), jnp.sqrt(2)]), shape=(k1,), p=jnp.array([1/5, 1/5, 1/5, 1/5, 1/5]))


# sort the vector v_teacher
v_teacher = np.sort(v_teacher)
v_sample = np.copy(v_teacher)

# Generate synthetic data
X = jax.random.normal(key_data, (n, d), dtype=jnp.float64)  # Inputs

def teacher_forward(x):
    h1 = activation((1/jnp.sqrt(d)) * x @ W1_teacher.T)
    #h2 = activation((1/jnp.sqrt(k1)) * h1 @ W2_teacher.T)
    return (1/jnp.sqrt(k1)) * h1 @ v_teacher

y = teacher_forward(X) + sigma * jax.random.normal(key_data, (n,), dtype=jnp.float64)

# Bayesian model with fixed readout (v)
def model(X, y=None):
    # Priors for weights (excluding readout)
    W1 = numpyro.sample("W1", dist.Normal(0, 1).expand([k1, d]).to_event(2))
    
    # Forward pass with fixed v_teacher
    h1 = activation((1/jnp.sqrt(d)) * X @ W1.T)
    #h2 = activation((1/jnp.sqrt(k1)) * h1 @ W2.T)
    f = (1/jnp.sqrt(k1)) * h1 @ v_teacher
    
    # Likelihood
    with numpyro.plate("obs", X.shape[0]):
        numpyro.sample("y", dist.Normal(f, sigma), obs=y)

if info_init:
    init_params = {"W1": W1_teacher}
else:
    init_params = {"W1": np.random.randn(k1, d)}

# Run NUTS sampler (adjust num_samples/warmup as needed)
nuts_kernel = NUTS(model)
mcmc = MCMC(nuts_kernel, num_warmup=num_warmup_samples, num_samples=num_posterior_samples)
mcmc.warmup(rng_key, X, y=y, init_params=init_params, collect_warmup=True)
#mcmc.run(rng_key, X, y=y)
#mcmc.print_summary()
```

```python
import matplotlib.pyplot as plt

# Extract posterior samples (W1 and W2)
samples = mcmc.get_samples()

samples_W1 = samples['W1']
#samples_W2 = samples['W2']
#samples_v = samples['v']

# ----- Compute teacher quantities -----
# S_1^star = (1/sqrt(k1)) * (W1_teacher.T @ (v_teacher))
S1_teacher = (1 / jnp.sqrt(k1)) * (W1_teacher.T @ (v_teacher))  # shape (d,)

# S_2^{(1)â‹†} = (1/k1) * (W1_teacher.T @ diag((1/sqrt(k2))*(W2_teacher.T @ v_teacher)) @ W1_teacher)
#temp_teacher = (1 / jnp.sqrt(k2)) * (W2_teacher.T @ v_teacher)  # shape (k1,)
S2_teacher = (1 / jnp.sqrt(k1)) * (W1_teacher.T @ jnp.diag(v_teacher) @ W1_teacher)  # shape (d, d)

# S_2^{(2)â‹†} = (1/(k1*sqrt(k2))) * (W1_teacher.T @ (W2_teacher.T@ diag(v_teacher) @ W2_teacher) @ W1_teacher)))
#S2_teacher_2 = (1 / (k1 * jnp.sqrt(k2))) * (W1_teacher.T @ (W2_teacher.T @ jnp.diag(v_teacher) @ W2_teacher) @ W1_teacher)  # shape (d, d)

const_QW_1 = (1 / (k1 * d)) * jnp.trace(W1_teacher @ W1_teacher.T)
#const_QW_2 = (1 / (k2 * k1)) * jnp.trace(W2_teacher @ W2_teacher.T)
const_Qv = (1/k1) * jnp.dot(v_teacher,v_teacher)
const_Q1 = (1 / d) * jnp.dot(S1_teacher, S1_teacher)
const_Q2 = (1 / (d * d)) * jnp.trace(S2_teacher @ S2_teacher.T)
const_Q3 = (v_teacher[None, :] @ (W1_teacher @ W1_teacher.T)**3 @ v_teacher[:, None])/(d**3 * k1)
const_Q4 = (v_teacher[None, :] @ (W1_teacher @ W1_teacher.T)**4 @ v_teacher[:, None])/((d**4) * k1)
#const_Q2_2 = (1 / (d * d)) * jnp.trace(S2_teacher_2 @ S2_teacher_2.T)

print('Teacher-teacher overlap Q_W^(1):', const_QW_1)
#print('Teacher-teacher overlap Q_W^(2):', const_QW_2)
print('Teacher-teacher overlap Q_v:', const_Qv)
print('Teacher-teacher overlap Q_1:', const_Q1)
print('Teacher-teacher overlap Q_2^(1):', const_Q2)
print('Teacher-teacher overlap Q_3:', const_Q3)
print('Teacher-teacher overlap Q_4:', const_Q4)
#print('Teacher-teacher overlap Q_2^(2):', const_Q2_2)

#const_QW_1 = 1.0
#const_QW_2 = 1.0
#const_Qv = 1.0
#const_Q1 = 1.0
#const_Q2_1 = 1.0
#const_Q2_2 = 1.0

# ----- Loop over posterior samples and compute order parameters -----
num_samples = samples_W1.shape[0]

# Preallocate lists for the overlaps over samples
QW1_list = []   # Q_W^(1)
#QW2_list = []   # Q_W^(2)
Qv_list = []    # Q_v
Q1_list  = []   # Q_1
Q2_list = []  # Q_2^(1)
#Q2_2_list = []  # Q_2^(2)
Q3_list = []  # Q_3
Q4_list = []  # Q_4

for i in range(num_samples):
    # Get the student parameters for the i-th sample.
    W1_sample = samples_W1[i]  # shape (k1, d)
    #W2_sample = samples_W2[i]  # shape (k2, k1)
    #v_sample = samples_v[i]    # shape (k1)
    #W1_sample, perm_W1 = align_matrix(W1_teacher, W1_sample, axis='row')
    #v_sample = v_sample[perm_W1]

    
    # Compute Q_W^(1) = (1/(k1*d)) * trace(W1_sample @ W1_teacher.T)
    QW1 = (1 / (k1 * d)) * jnp.trace(W1_sample @ W1_teacher.T)
    QW1_list.append(float(QW1)/const_QW_1)
    
    # Compute Q_W^(2) = (1/(k2*k1)) * trace(W2_sample @ W2_teacher.T)
    #QW2 = (1 / (k2 * k1)) * jnp.trace(W2_sample @ W2_teacher.T)
    #QW2_list.append(float(QW2)/const_QW_2)
    
    # Compute Q_v = (1/k2) * dot(v_sample,v_teacher)
    Qv = (1/k1) * jnp.dot(v_sample,v_teacher)
    Qv_list.append(float(Qv)/const_Qv)
    
    # Compute S_1 for student:
    # S_1 = (1/sqrt(k1*k2)) * (W1_sample.T @ ( v_sample))
    S1_sample = (1 / jnp.sqrt(k1)) * (W1_sample.T @ (v_sample))
    # Then Q_1 = (1/d)* dot(S1_teacher, S1_sample)
    Q1_val = (1 / d) * jnp.dot(S1_teacher, S1_sample)
    Q1_list.append(float(Q1_val)/const_Q1)
    
    # Compute S_2^(1) for student:
    # S_2^(1) = (1/sqrt(k1)) * (W1_sample.T @ diag((1/sqrt(k2))*(W2_sample.T @ v_sample)) @ W1_sample)
    #temp_sample = (1 / jnp.sqrt(k2)) * (W2_sample.T @ v_sample)  # shape (k1,)
    S2_sample = (1 / jnp.sqrt(k1)) * (W1_sample.T @ jnp.diag(v_sample) @ W1_sample)
    # Then Q_2^(1) = (1/(d^2)) * trace( S2_sample_1 @ S2_teacher_1.T )
    Q2 = (1 / (d * d)) * jnp.trace(S2_sample @ S2_teacher.T)
    Q2_list.append(float(Q2)/const_Q2)
    
    # Compute S_2^(2) for student:
    # S_2^(2) = (1/(k1*sqrt(k2))) * (W1_sample.T @ (W2_sample.T @ diag(v_sample) @ W2_sample) @ W1_sample)))
    #S2_sample_2 = (1 / (k1 * jnp.sqrt(k2))) * (W1_sample.T @ (W2_sample.T @ jnp.diag(v_sample) @ W2_sample) @ W1_sample)
    # Then Q_2^(2) = (1/(d^2)) * trace( S2_sample_2 @ S2_teacher_2.T )
    #Q2_2 = (1 / (d * d)) * jnp.trace(S2_sample_2 @ S2_teacher_2.T)
    #Q2_2_list.append(float(Q2_2)/const_Q2_2)

    # Compute Q_3 sample-teacher overlap
    Q3 = (v_sample[None, :] @ (W1_sample @ W1_teacher.T)**3 @ v_teacher[:, None])/(d**3 * k1)
    Q3 = Q3 / const_Q3
    Q3_list.append(float(Q3.item()))

    # Compute Q_4 sample-teacher overlap
    Q4 = (v_sample[None, :] @ (W1_sample @ W1_teacher.T)**4 @ v_teacher[:, None])/(d**4 * k1)
    Q4 = Q4 / const_Q4
    Q4_list.append(float(Q4.item()))

# ----- Plot the order parameters as a function of sample index -----
iterations = np.arange(1, num_samples + 1)
plt.figure(figsize=(12, 10))
plt.plot(iterations, np.array(QW1_list)**1, label=r'$Q_W$', linewidth=2)
#plt.plot(iterations, QW2_list, label=r'$Q_W^{(2)}$', linewidth=2)
#plt.plot(iterations, Qv_list, label=r'$Q_v$', linewidth=2)
plt.plot(iterations, Q1_list,  label=r'$Q_1$', linewidth=2)
plt.plot(iterations, Q2_list, label=r'$Q_2$', linewidth=2)
#plt.plot(iterations, Q2_2_list, label=r'$Q_2^{(2)}$', linewidth=2)
plt.plot(iterations, np.array(Q3_list), label=r'$Q_3$', linewidth=2)
plt.plot(iterations, np.array(Q4_list), label=r'$Q_4$', linewidth=2)
plt.xlabel("Sample Index")
plt.ylabel("Overlap")
plt.title("Hermite 1,2,3 with alpha="+str(alpha))
plt.legend()
plt.tight_layout()
#plt.savefig('overlaps.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
from scipy.optimize import linear_sum_assignment

def permute(A, B):
    """
    Permutes the rows of A such that the trace of (A_permuted * B^T) is maximized.
    
    That is, finds a permutation Ïƒ such that:
      trace(A_permuted * B^T) = sum_{j=0}^{k-1} dot(A[Ïƒ(j),:], B[j,:])
    is maximized.
    
    Parameters:
        A (np.ndarray): A matrix of shape (k, d).
        B (np.ndarray): A matrix of shape (k, d).
    
    Returns:
        perm (np.ndarray): An array of length k where perm[j] gives the row index of A 
                           that should be placed in the j-th row of A_permuted.
        max_trace (float): The maximum trace value achieved by this permutation.
    """
    # Compute the cost matrix where cost[i, j] is the dot product of A[i] and B[j]
    cost_matrix = np.dot(A, B.T)
    
    # To maximize the trace, we minimize the negative of the cost matrix.
    row_ind, col_ind = linear_sum_assignment(-cost_matrix)
    
    # The assignment from the Hungarian algorithm gives pairs (i, j) where:
    # row_ind[i] is a row from A assigned to column (i.e., row) col_ind[i] of B.
    # We need a permutation array perm such that:
    #   A_permuted[j, :] = A[perm[j], :]
    # where j corresponds to B's row.
    perm = np.empty_like(col_ind)
    for r, c in zip(row_ind, col_ind):
        perm[c] = r  # For the B row c, choose A row r.
    
    # Compute the maximum trace using the computed permutation
    # That is, sum_{j} dot( A[perm[j], :], B[j, :] )
    max_trace = sum(np.dot(A[perm[j], :], B[j, :]) for j in range(len(perm)))
    
    return perm, max_trace
```
