```python
import numpy as np

import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt

import sys
import os
folder = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.insert(0, folder)

import theory2
```

```python
gamma = 0.5
Delta = 1.25
vlaw = 'radd'
prior= 'gauss'


sig = lambda x: x + (x**2 - 1)/np.sqrt(2.0) + (x**3 - 3*x)/6
sp_quantities = theory2.sp_quantities(sig, 5, gamma, Delta, vlaw, prior)
uni_quantities = theory2.uni_quantities(vlaw, sig, 0.5, gamma, Delta)
```

```python
colors = ["#FF0000", "#C40000", "#800000", "#330000"]
```

```python
width, height = 6, 5
big, small = int(width*3), int(width*2.5)
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(width, height), tight_layout=True)

# upper fig
data_uni = np.loadtxt('data_uni.csv', delimiter=',')
q1s, q2s, q3s, q4s, q5s, qws = data_uni

ax1.plot(q1s, label=r'$q_1$', color='g')
ax1.plot(q2s, label=r'$q_2$', color='b')
ax1.plot(q3s, label=r'$q_3$', color=colors[0])
ax1.plot(q4s, label=r'$q_4$', color=colors[1])
ax1.plot(q5s, label=r'$q_5$', color=colors[2])
ax1.plot(qws, label=r'$q_W$', color=colors[3])

ax1.axhline(y=uni_quantities['q2'], color='b', linestyle='--')
ax1.axhline(y=0, color='black', linestyle='--')

ax1.set_xlim(0, 2000)
ax1.tick_params(axis='both', labelsize=small)
ax1.legend(fontsize=13, loc="upper right")

# lower fig
data_sp = np.loadtxt('data_sp.csv', delimiter=',')
q1s, q2s, q3s, q4s, q5s, qws = data_sp
ax2.plot(q1s, label=r'$q_1$', color='g')
ax2.plot(q2s**(1/2), label=r'$q_2^{1/2}$', color='b')
ax2.plot(q3s**(1/3), label=r'$q_3^{1/3}$', color=colors[0])
ax2.plot(q4s**(1/4), label=r'$q_4^{1/4}$', color=colors[1])
ax2.plot(q5s**(1/5), label=r'$q_5^{1/5}$', color=colors[2])
ax2.plot(qws, label=r'$q_W$', color=colors[3])

ax2.axhline(y=sp_quantities['q2']**(1/2), color='b', linestyle='--')
ax2.axhline(y=sp_quantities['qw'], color='black', linestyle='--')

ax2.set_xlim(0, 2000)
ax2.tick_params(axis='both', labelsize=small)
ax2.legend(fontsize=12, loc="upper right")
ax2.set_xlabel('HMC step', fontsize=small)

fig.tight_layout(rect=[0.02, 0, 1, 1])
plt.savefig('overlaps.pdf')
```
