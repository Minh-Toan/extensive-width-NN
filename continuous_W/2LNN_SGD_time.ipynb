```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import math
import time
import csv
import os

# Use GPU if available
device = torch.device("cpu" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ---------------------
# Experiment Parameters
# ---------------------
alpha = 5              # so that n = alpha * d^2
gamma = 0.5            # so that k = gamma * d
Delta = 0.0001         # noise variance
threshold = 0.01      # test loss threshold
num_teachers = 20      # number of independent teacher runs per dimension
num_epochs = 5000      # maximum epochs per candidate run
big_value = 1e6        # value if threshold is not reached
# Dimensions to test:
d_vals = [260]
# Fixed learning rate (for all experiments)
learning_rate = 0.05

# Readout type: "gaussian", "rademacher", "uniform", "ones"
readout_type = "gaussian"  # Change this to switch readout type

# CSV file name
csv_filename = "metrics_gauss_001.csv"

# If file doesn't exist, write header first.
if not os.path.exists(csv_filename):
    with open(csv_filename, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["dimension", "avg_min_time (s)", "std_min_time (s)", 
                         "avg_min_grad_updates", "std_min_grad_updates"])

# ---------------------
# Define Student Model
# ---------------------
class Student(nn.Module):
    def __init__(self, d, k, readout):
        super().__init__()
        self.fc1 = nn.Linear(d, k, bias=False)
        with torch.no_grad():
            self.fc1.weight.div_(math.sqrt(d))
        self.register_buffer('readout', readout)
    
    def forward(self, x):
        h = torch.relu(self.fc1(x))
        return h @ self.readout / math.sqrt(h.size(1))

# ---------------------
# Loop over dimensions
# ---------------------
for d in d_vals:
    print(f"\n=== Dimension d = {d} ===")
    k = int(gamma * d)
    n = int(alpha * d**2)
    
    L = int(math.floor(np.log2(n)))
    candidate_bs = sorted([max(1, int(n / (2**p))) for p in range(2, L-1)], reverse=True)
    print("Candidate batch sizes:", candidate_bs)
    
    teacher_time = []
    teacher_grad_updates = []
    # For each candidate batch size, we store a list of times recorded from previous teacher runs.
    batch_performance = {bs: [] for bs in candidate_bs}
    
    for t_run in range(num_teachers):
        W_teacher = torch.randn(d, k, device=device) / math.sqrt(d)
        
        if readout_type == "gaussian":
            readout = torch.randn(k, 1, device=device)
        elif readout_type == "rademacher":
            readout = torch.randint(0, 2, (k, 1), device=device, dtype=torch.float32) * 2 - 1
        elif readout_type == "uniform":
            a = math.sqrt(3)
            readout = (torch.rand(k, 1, device=device) * (2 * a) - a)
        elif readout_type == "ones":
            readout = torch.ones(k, 1, device=device)
        else:
            raise ValueError("Invalid readout_type. Choose from 'gaussian', 'rademacher', 'uniform', or 'ones'.")
        
        X_train = torch.randn(n, d, device=device)
        hidden_train = torch.relu(X_train @ W_teacher)
        Y_train = hidden_train @ readout / math.sqrt(k)
        Y_train += torch.randn(n, 1, device=device) * (Delta**0.5)
        
        X_test = torch.randn(10000, d, device=device)
        hidden_test = torch.relu(X_test @ W_teacher)
        Y_test = hidden_test @ readout / math.sqrt(k)
        Y_test += torch.randn(10000, 1, device=device) * (Delta**0.5)
        
        best_time = big_value
        best_grad = big_value
        
        # Reorder candidate batch sizes based on their best (minimum) recorded performance so far
        candidate_bs.sort(key=lambda bs: min(batch_performance[bs]) if batch_performance[bs] else big_value)
        
        for bs in candidate_bs:
            student = Student(d, k, readout).to(device)
            optimizer = optim.Adam(student.parameters(), lr=learning_rate)
            criterion = nn.MSELoss()
            
            grad_updates = 0
            candidate_time = big_value
            candidate_grad = big_value
            
            start_time = time.time()
            loss_window = []
            reached = False
            
            for epoch in range(num_epochs):
                student.train()
                permutation = torch.randperm(n, device=device)
                for i in range(0, n, bs):
                    indices = permutation[i:i+bs]
                    batch_X = X_train[indices]
                    batch_Y = Y_train[indices]
                    
                    optimizer.zero_grad()
                    outputs = student(batch_X)
                    loss = criterion(outputs, batch_Y)
                    loss.backward()
                    optimizer.step()
                    
                    grad_updates += 1
                    student.eval()
                    with torch.no_grad():
                        current_test_loss = criterion(student(X_test), Y_test).item()
                    
                    loss_window.append(current_test_loss)
                    if len(loss_window) > 10:
                        loss_window.pop(0)
                    
                    if len(loss_window) == 10 and sum(loss_window) / 10 < threshold:
                        candidate_time = time.time() - start_time
                        candidate_grad = grad_updates
                        reached = True
                        break
                if reached:
                    break
                
                if time.time() - start_time > best_time or grad_updates > best_grad:
                    break
            
            if candidate_time < best_time:
                best_time = candidate_time
                batch_performance[bs].append(best_time)
            else:
                batch_performance[bs].append(big_value)
            
            best_grad = min(best_grad, candidate_grad)
        
        teacher_time.append(best_time)
        teacher_grad_updates.append(best_grad)
        print(f"Teacher run {t_run+1}: time = {best_time:.2f} s, grad updates = {best_grad}")
    
    avg_time = np.mean(teacher_time)
    std_time = np.std(teacher_time)
    avg_grad = np.mean(teacher_grad_updates)
    std_grad = np.std(teacher_grad_updates)
    
    print(f"Dimension {d}: avg_time = {avg_time:.2f} s (std = {std_time:.2f}), avg_grad_updates = {avg_grad:.2f} (std = {std_grad:.2f})")
    
    with open(csv_filename, "a", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow([d, avg_time, std_time, avg_grad, std_grad])

print(f"Results appended to {csv_filename}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import math
import time
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import csv  # For saving CSV files

# Use GPU if available
device = torch.device("cpu" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ---------------------
# Experiment Parameters
# ---------------------
d_vals = [60, 80, 100, 120, 140, 160, 180]  # Dimensions to test
alpha = 5                    # so that n = alpha * d^2
gamma = 0.5                  # so that k = gamma * d
Delta = 0.0001               # noise variance
learning_rate = 0.05         # fixed learning rate
num_epochs = 5000            # maximum epochs per teacher run
max_grad_steps = 6000        # maximum gradient steps allowed
num_teachers = 5             # number of teacher runs per dimension

# Readout type: only "gaussian" is used for this experiment
readout_type = "ones"

# ---------------------
# Derived Parameters and Model Definition
# ---------------------
def get_readout(k, readout_type):
    if readout_type == "gaussian":
        return torch.randn(k, 1, device=device)
    elif readout_type == "rademacher":
        return torch.randint(0, 2, (k, 1), device=device, dtype=torch.float32) * 2 - 1
    elif readout_type == "uniform":
        a = math.sqrt(3)
        return torch.rand(k, 1, device=device) * (2 * a) - a
    elif readout_type == "ones":
        return torch.ones(k, 1, device=device)
    else:
        raise ValueError("Invalid readout_type. Choose from 'gaussian', 'rademacher', 'uniform', or 'ones'.")

class Student(nn.Module):
    def __init__(self, d, k, readout):
        super().__init__()
        self.fc1 = nn.Linear(d, k, bias=False)
        with torch.no_grad():
            self.fc1.weight.div_(math.sqrt(d))
        self.register_buffer('readout', readout)
    
    def forward(self, x):
        h = torch.relu(self.fc1(x))
        return h @ self.readout / math.sqrt(h.size(1))

# ---------------------
# Prepare CSV storage
# ---------------------
# We'll create a list of rows, where each row contains: dimension, gradient_step, mean_loss, std_loss
csv_rows = []
csv_rows.append(["d", "gradient_step", "mean_loss", "std_loss"])

# ---------------------
# Plotting Phase: Average Over 5 Teachers for Each Dimension
# ---------------------
plt.figure(figsize=(10, 6))
small = 12  # Font size for labels

# Define color gradient from red to black
colors = [cm.Reds(i) for i in np.linspace(0.3, 1, len(d_vals))]  # Reds colormap

for d, color in zip(d_vals, colors):
    k = int(gamma * d)
    n = int(alpha * d**2)
    batch_size = max(1, int(n / 4))  # Fixed batch size = n/4
    
    print(f"\nDimension d = {d}: using batch size = {batch_size}")
    
    all_test_losses = []

    for t_run in range(num_teachers):
        # Generate teacher data with a Gaussian readout
        W_teacher = torch.randn(d, k, device=device) / math.sqrt(d)
        teacher_readout = get_readout(k, readout_type)
        
        X_train = torch.randn(n, d, device=device)
        hidden_train = torch.relu(X_train @ W_teacher)
        Y_train = hidden_train @ teacher_readout / math.sqrt(k)
        Y_train += torch.randn(n, 1, device=device) * (Delta**0.5)
        
        X_test = torch.randn(10000, d, device=device)
        hidden_test = torch.relu(X_test @ W_teacher)
        Y_test = hidden_test @ teacher_readout / math.sqrt(k)
        Y_test += torch.randn(10000, 1, device=device) * (Delta**0.5)
        
        # Initialize student with the same readout as teacher
        student = Student(d, k, teacher_readout).to(device)
        optimizer = optim.Adam(student.parameters(), lr=learning_rate)
        criterion = nn.MSELoss()
        
        test_losses = []
        grad_steps = 0
        
        for epoch in range(num_epochs):
            student.train()
            permutation = torch.randperm(n, device=device)
            for i in range(0, n, batch_size):
                indices = permutation[i:i+batch_size]
                batch_X = X_train[indices]
                batch_Y = Y_train[indices]
                
                optimizer.zero_grad()
                outputs = student(batch_X)
                loss = criterion(outputs, batch_Y)
                loss.backward()
                optimizer.step()
                grad_steps += 1
                
                student.eval()
                with torch.no_grad():
                    current_loss = criterion(student(X_test), Y_test).item()
                test_losses.append(current_loss)
                
                if grad_steps > max_grad_steps:
                    break
            if grad_steps > max_grad_steps:
                break
        
        all_test_losses.append(test_losses)

    # Pad all test loss sequences to the same length (for averaging)
    max_length = max(len(tl) for tl in all_test_losses)
    padded_losses = [np.pad(tl, (0, max_length - len(tl)), constant_values=tl[-1]) for tl in all_test_losses]
    
    mean_loss = np.mean(padded_losses, axis=0)
    std_loss = np.std(padded_losses, axis=0)

    # Save the mean and std for each gradient step for this dimension to CSV rows
    for grad_step in range(len(mean_loss)):
        csv_rows.append([d, grad_step, mean_loss[grad_step], std_loss[grad_step]])

    # Plot mean test loss with shaded variance
    x_axis = np.arange(len(mean_loss))
    plt.plot(x_axis, mean_loss, color=color, label=f"d = {d}")
    plt.fill_between(x_axis, mean_loss - std_loss, mean_loss + std_loss, color=color, alpha=0.2)

# Write CSV file after processing all dimensions
with open("mean_test_losses_ones.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(csv_rows)

# Plot threshold lines with labels
plt.axhline(y=0.0242, color='black', linestyle='dashed', label=r'$2 \, \varepsilon^{\rm uni}$')
plt.axhline(y=0.0121, color='black', linestyle='dashed', label=r'$\varepsilon^{\rm uni}$')
plt.axhline(y=1e-5, color='red', linestyle='dashed', label=r'$\varepsilon^{\rm sp}$')

plt.xlabel("Gradient updates", fontsize=small)
plt.ylabel("Test loss", fontsize=small)
plt.grid(True, which='both')
plt.tight_layout()
plt.legend(fontsize=small)
plt.tick_params(axis='both', which='major', labelsize=small)
plt.tick_params(axis='both', which='minor', labelsize=small)
plt.ylim((0, 0.1))
#plt.savefig('test_loss_gauss.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()
```
