```python
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import csv

import sys
import os

folder = os.path.abspath(os.path.join(os.getcwd()))
```

```python
def csv_to_dict(filename):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        data = {header: [] for header in reader.fieldnames}
        for row in reader:
            for header in reader.fieldnames:
                data[header].append(float(row[header]))
        data = {key: np.array(value) for key, value in data.items()}
    return data
```

```python
gamma = 0.5
Delta = 1.25
activation = 'h23'
vlaw = 'ones'
prior= 'gauss'
path_theory = folder + '/data_theory/'
path_exp =  folder + fr'/{activation}/v_{vlaw}/'

data_uni = csv_to_dict(path_theory + fr'{activation}_Delta_{Delta}_gamma_{gamma}_v_{vlaw}_uni.csv')
data_spe = csv_to_dict(path_theory + fr'{activation}_Delta_{Delta}_gamma_{gamma}_v_{vlaw}_sp.csv')
```

```python
width, height = 6, 5
big, small = int(width*3), int(width*2.5)
fig, ax = plt.subplots(figsize=(width, height), tight_layout=True)

# theory
plt.plot(data_uni['alpha'], data_uni['mmse'], label='universal', color='blue')
plt.plot(data_spe['alpha'], data_spe['mmse'], label='specialisation', color='red')
    
# experiment
stats = csv_to_dict(path_exp + 'stats.csv')
plt.scatter(stats['alpha'], stats['info_mean'], marker='o', edgecolors='black', facecolors='none', s=20, label='informative HMC')
plt.scatter(stats['alpha'], stats['unfo_mean'], marker='x', color='black', s=20, label='uninformative HMC')
plt.tick_params(axis='both', labelsize=big)


plt.xlabel(r'$\alpha$', fontsize=big)
plt.ylabel('Generalization error', fontsize=big)
plt.legend(fontsize=small)
plt.xlim(0, 8.1)
# plt.ylim(0, 0.04)
plt.title(fr'$\gamma={gamma}, \Delta={Delta}, v$ {vlaw}', fontsize=big)
# plt.title(fr'ReLU, $\gamma={gamma}, \Delta={Delta}, v$ {vlaw}', fontsize=big)
plt.show()
```
